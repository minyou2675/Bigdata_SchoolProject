```python
import re # regular expression
str = '우리가 만약 Tom과 만나게 되면 Jane은 뭐라고 할까? 윌리엄은? 그러면 picnic은? 그럼 K는?'
# 영어로 된 사람 이름: 하나의(첫 문자) 대문자, 한 개 이상의 소문자가 나온다.

re.findall("[A-Z][a-z]+", str)
```

```python
str2 = "네가 185번 버스를 타고 가다가 130번으로 갈아타면 될 것 같은데, 비용은 1,500원이 들 거야."

# \d는 숫자라는 뜻
re.findall('[\d]+', str2)
```

```python
# 1하고 500을 다로 찾지 않으려면
re.findall('[\d]+,?[\d]*', str2)
```

```python
re.findall('[\d]+,[\d]+|[\d]+', str2)
```

```python
# 100만은 어떻게 할까?
re.findall('[\d].*?[\d]*', str2)
```

```python
re_iter = re.finditer('[A-Z][a-z]+', str) # type: callable_iterator
for s in re_iter:
  print(s)
```

```python
re.findall(r'<td class="left">.*?td>', text, re.DOTALL)
```

```python
re.findall(r'<td.*?td>', text, re.DOTALL)
```

```python
re.findall(r'<td.*?/td?>', text, re.DOTALL)
```

```python
urls =[]
titles = []
authors =[]
dates = []
prices = []
brands = []

for item in re.findall(r'<td.*?/td?>', text, re.DOTALL):
  if(re.findall(r'<td class="left">.*?/td>', item, re.DOTALL)):
    res = re.search(r'<a href="(.*?)">', item)
    if res != None:
      url = re.search(r'<a href="(.*?)">', item).group(1)
      title = re.sub(r'<.*?>', '', item)
      print("url:", url)
      print("title:", title)
      urls.append(url)
      titles.append(title)
    else:
      author = re.sub(r'<.*?>', '', item)
      print("author:", author)
      authors.append(author)

  if (re.findall(r'<td>.*?/td>', item, re.DOTALL)):
    date = re.sub(r'<.*?>', '', item)
    date = unescape(date)
    print("date:", date)
    dates.append(date)

  if (re.findall(r'<td class="right">.*/td>', item, re.DOTALL)):
    price = re.sub(r'<.*?>', '', item)
    price = unescape(price)
    print("price:", price)
    prices.append(price)

  if (re.findall(r'<td class="brd_.*">.*/td>', item, re.DOTALL)):
    brand = re.sub(r'<.*?>', '', item)
    brand = unescape(brand)
    print("brand:", brand)
    brands.append(brand)
```

```python
import pandas as pd

df = pd.DataFrame (
    {
      "URL": urls,
      "TITLE": titles,
     "AUTHOR": authors,
     "DATE": dates,
     "PRICE": prices,
     "BRAND": brands
    }
)
```
