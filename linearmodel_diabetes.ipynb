```python
from sklearn.model_selection import train_test_split #sk에서 제공해주는 비율 나누는 함수
from sklearn.linear_model import LinearRegression #loss,쎄타 함수 내장 

import pandas as pd
import numpy as np
```

```python
from sklearn.datasets import load_diabetes
diabetes = load_diabetes()
x, y = diabetes["data"], diabetes["target"]
```

```python
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size =0.2) #데이터셋은 랜덤하게 분리된다
```

```python
len(x),len(train_x), len(test_x), len(train_y), len(test_y)
```

```python
regressor = LinearRegression ()
regressor.fit(train_x, train_y)
```

```python
regressor.coef_ #na값 찍히면 잘못 된 것
```

```python
train_pred = regressor.predict(train_x)
test_pred = regressor.predict(test_x)
```

```python
print(train_pred[1],train_y[1])
```

```python
from sklearn.metrics import mean_squared_error
```

```python
train_mse = mean_squared_error(train_pred, train_y)
test_mse = mean_squared_error(test_pred, test_y)
print(train_mse, test_mse) 
```
